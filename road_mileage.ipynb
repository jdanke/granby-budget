```python
import PyPDF2 as pypdf
import re
```

```python
# https://portal.ct.gov/dot/-/media/dot/policy/publicroad/publicroadmileage_final.pdf
with open('publicroadmileage_final.pdf', 'rb') as file:
    reader = pypdf.PdfReader(file)
    num_pages = len(reader.pages)
    print(f"Number of pages: {num_pages}")

    for page in reader.pages:
        text = page.extract_text()
        print('* ', text, ' *')
```
